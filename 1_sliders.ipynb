```python
import pathlib
import numpy as np  # TODO: tell them to install numpy and matplotlib
import matplotlib.pyplot as plt
from random import uniform

from ipywidgets import interact, FloatSlider
from IPython.display import display
```

```python
DATA_PATH = pathlib.Path("data/sliders")
```

```python
def interact_classification(data: np.array):
    low = data[:, :2].min()
    high = data[:, :2].max()
    fig = plt.figure(figsize=(6, 6))

    line, = plt.plot([], [], color='green', label="Classifier")
    A = data[data[:, 2] == 0]
    B = data[data[:, 2] == 1]
    plt.scatter(A[:, 0], A[:, 1], color='b', label="Class A");
    plt.scatter(B[:, 0], B[:, 1], color='red', label="Class B");
    plt.legend()
    plt.close()
  
    def plot_classifier(w, b):
        line.set_data((low, high), [w * i + b for i in (low, high)])
        display(fig)


    interact(plot_classifier, w=FloatSlider(min=-5, max=5, step=0.1, value=0.0), 
                              b=FloatSlider(min=-3, max=3, step=0.1, value=0.0))
    print(type(fig))
```

```python
interact_classification(np.load(DATA_PATH / "clear.npy"))
```

```python
interact_classification(np.load(DATA_PATH / "fuzzy.npy"))
```
